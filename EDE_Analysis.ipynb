#**Review Analysis**

```python
import re
import nltk
import pandas as pd
import seaborn as sns
nltk.download('vader_lexicon')
import matplotlib.pyplot as plt
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
fb['row_id'] = fb.index + 1
fb.head()
```

```python
fb_subset = fb[['row_id', 'Review']].copy()
```

```python
# removing all non-aphabet characters
fb_subset['Review'] = fb_subset['Review'].str.replace("[^a-zA-Z#]", " ")
```

```python
# convert to lower-case
fb_subset['Review'] = fb_subset['Review'].str.casefold()
print (fb_subset.head(10))
```

```python
# set up empty dataframe for staging output
fb1=pd.DataFrame()
fb1['row_id'] = ['99999999999']
fb1['sentiment_type'] = 'NA999NA'
fb1['sentiment_score'] = 0
```

```python
sid = SentimentIntensityAnalyzer()
t_fb = fb1
for index, row in fb_subset.iterrows():
    scores = sid.polarity_scores(row[1])
    for key, value in scores.items():
        temp = [key,value,row[0]]
        fb1['row_id']=row[0]
        fb1['sentiment_type']=key
        fb1['sentiment_score']=value
        t_fb=t_fb.append(fb1)
```

```python
# remove dummy row with row_id = 99999999999
t_fb_cleaned = t_fb[t_fb.row_id != '99999999999']
```

```python
# remove duplicates if any exist
t_fb_cleaned = t_fb_cleaned.drop_duplicates()
```

```python
# only keep rows where sentiment_type = compound
t_fb_cleaned = t_fb[t_fb.sentiment_type == 'compound']
print(t_fb_cleaned.head(10))
```

```python
# merge dataframes
fb_output = pd.merge(fb, t_fb_cleaned, on='row_id', how='inner')
```
